```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'Tweets.csv'
social_media_data = pd.read_csv(file_path)


print(social_media_data.head())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='airline_sentiment', data=social_media_data)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='airline_sentiment_confidence', data=social_media_data, bins=20, kde=True)
plt.title('Sentiment Confidence Distribution')
plt.xlabel('Sentiment Confidence')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='negativereason', data=social_media_data, order=social_media_data['negativereason'].value_counts().index)
plt.title('Negative Reasons Distribution')
plt.xlabel('Negative Reason')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='airline', hue='airline_sentiment', data=social_media_data)
plt.title('Airline-wise Sentiment Distribution')
plt.xlabel('Airline')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='airline', y='airline_sentiment_confidence', data=social_media_data)
plt.title('Sentiment Confidence by Airline')
plt.xlabel('Airline')
plt.ylabel('Sentiment Confidence')
plt.show()
```

```python
social_media_data['tweet_length'] = social_media_data['text'].apply(len)
plt.figure(figsize=(12, 8))
sns.boxplot(x='airline_sentiment', y='tweet_length', data=social_media_data)
plt.title('Tweet Length vs. Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Tweet Length')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='user_timezone', hue='airline_sentiment', data=social_media_data, order=social_media_data['user_timezone'].value_counts().index)
plt.title('Sentiment Distribution by User Timezone')
plt.xlabel('User Timezone')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='airline_sentiment', y='retweet_count', data=social_media_data)
plt.title('Retweet Count vs. Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Retweet Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='negativereason', data=social_media_data, order=social_media_data['negativereason'].value_counts().index)
plt.title('Distribution of Negative Reasons')
plt.show()
```

```python
social_media_data['word_count'] = social_media_data['text'].apply(lambda x: len(str(x).split()))
plt.figure(figsize=(12, 8))
sns.histplot(x='word_count', hue='airline_sentiment', data=social_media_data, bins=30, kde=True)
plt.title('Sentiment Word Count Distribution')
plt.show()
```

```python
sentiment_counts = social_media_data['airline_sentiment'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Sentiment Pie Chart')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='airline', hue='negativereason', data=social_media_data)
plt.title('Negative Reason Distribution by Airline')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```
