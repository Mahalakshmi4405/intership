```python
plt.figure(figsize=(10, 6))
sns.histplot(house_data['price'], bins=30, kde=True)
plt.title('House Price Distribution')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=house_data)
plt.title('Number of Bedrooms vs. House Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=house_data)
plt.title('Square Footage vs. House Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='view', y='price', data=house_data)
plt.title('View vs. House Price')
plt.xlabel('View')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='condition', y='price', data=house_data)
plt.title('Condition vs. House Price')
plt.xlabel('Condition')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='floors', y='price', data=house_data)
plt.title('Floors vs. House Price')
plt.xlabel('Floors')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.violinplot(x='zipcode', y='price', data=house_data)
plt.title('House Price Distribution by Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='yr_built', y='price', data=house_data)
plt.title('Year Built vs. House Price')
plt.xlabel('Year Built')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='yr_renovated', y='price', data=house_data)
plt.title('Renovation Impact on House Price')
plt.xlabel('Year Renovated')
plt.ylabel('Price')
plt.show()
```

```python
house_data['total_living_area'] = house_data['sqft_above'] + house_data['sqft_basement']
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_living_area', y='price', data=house_data)
plt.title('Total Living Area vs. House Price')
plt.xlabel('Total Living Area')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living15', y='price', data=house_data)
plt.title('Living Area in 2015 vs. House Price')
plt.xlabel('Living Area in 2015')
plt.ylabel('Price')
plt.show()
```

```python
correlation_matrix = house_data.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
