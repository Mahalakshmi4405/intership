```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = 'WA_Fn-UseC_-HR-Employee-Attrition.csv'
employee_data = pd.read_csv(file_path)


print(employee_data.head())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=employee_data)
plt.title('Attrition Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Attrition', y='Age', data=employee_data)
plt.title('Age Distribution by Attrition')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='BusinessTravel', hue='Attrition', data=employee_data)
plt.title('Business Travel vs. Attrition')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Department', hue='Attrition', data=employee_data)
plt.title('Department-wise Attrition')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=employee_data)
plt.title('Distance from Home vs. Attrition')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='JobRole', hue='Attrition', data=employee_data)
plt.title('Job Role vs. Attrition')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Attrition', y='MonthlyIncome', data=employee_data)
plt.title('Monthly Income Distribution by Attrition')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='OverTime', hue='Attrition', data=employee_data)
plt.title('Overtime vs. Attrition')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='PercentSalaryHike', hue='Attrition', data=employee_data, bins=20, kde=True)
plt.title('Percent Salary Hike Distribution by Attrition')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='YearsAtCompany', data=employee_data)
plt.title('Years at Company vs. Attrition')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='MaritalStatus', hue='Attrition', data=employee_data)
plt.title('Marital Status vs. Attrition')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='YearsInCurrentRole', data=employee_data)
plt.title('Years in Current Role vs. Attrition')
plt.show()
```

```python
categorical_features = employee_data.select_dtypes(include=['object']).columns
plt.figure(figsize=(16, 12))
for i, column in enumerate(categorical_features, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=column, y='Age', hue='Attrition', data=employee_data)
    plt.title(f'Boxplot for {column} vs. Age')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.kdeplot(x='JobSatisfaction', y='PerformanceRating', fill=True, data=employee_data)
plt.title('2D KDE for Job Satisfaction and Performance Rating')
plt.show()
```

```python
selected_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction']
sns.pairplot(employee_data[selected_features], diag_kind='kde')
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(employee_data, x='YearsAtCompany', y='MonthlyIncome', color='Attrition', hover_data=['JobRole'])
fig.update_layout(title='Interactive Scatter Plot: Years at Company vs. Monthly Income', xaxis_title='Years at Company', yaxis_title='Monthly Income')
fig.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='YearsInCurrentRole', y='Age', data=employee_data, inner='quartile')
sns.swarmplot(x='YearsInCurrentRole', y='Age', data=employee_data, color='black', alpha=0.5)
plt.title('Violin Plot with Swarm Plot: Years in Current Role and Age')
plt.show()
```
